```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
import dlib
```

```python
my_image_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/image.png' #     .
img_bgr = cv2.imread(my_image_path)    # OpenCV  
img_show = img_bgr.copy()      #    
plt.imshow(img_bgr) #      . (   )
plt.show() #  . ( )
```

```python
# plt.imshow  RGB    . 
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb) #       (        .)
plt.show() #   
```

    cv2.cvtColor   BGR2RGB 

```python
detector_hog = dlib.get_frontal_face_detector()
dlib_rects = detector_hog(img_rgb, 1)   #- (image, num of img pyramid)
print('   :', len(dlib_rects))
```

```python
# detector 
detector_hog = dlib.get_frontal_face_detector() #    
print("")
```

```python
#     
#     
print(dlib_rects)   

for dlib_rect in dlib_rects: #    
    l = dlib_rect.left() # 
    t = dlib_rect.top() # 
    r = dlib_rect.right() # 
    b = dlib_rect.bottom() # 

    cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA) #       

img_show_rgb =  cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

```python
model_path = os.getenv('HOME')+'/aiffel/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
    #  landmark   model_path  
landmark_predictor = dlib.shape_predictor(model_path)
    # dlib  shape_predictor    
    # landmark_predictor RGB dlib.rectangle   dlib.full_object_detection 
    # dlib.rectangle:   
    # dlib.full_object_detection:    ,     
print("")
```

```python
list_landmarks = []
    #    list     

#     face landmark 
# face landmark  
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
        #  landmark  points  
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
        #  landmark  (x,y)   list_points  
    list_landmarks.append(list_points)
        # list_landmarks   

print(len(list_landmarks[0]))
    #  n  list_landmarks n  
    #   68    list 
    # list_landmarks  1 list_landmarks[1]  IndexError 
```

```python
# list_landmarks  1    
for landmark in list_landmarks:
    for point in landmark:
        cv2.circle(img_show, point, 2, (0, 255, 255), -1)
            # cv2.circle: OpenCV   
            # img_show    point
            #  2 (0, 255, 255)  (-1)  
            # (       )

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
    # RGB  
plt.imshow(img_show_rgb)
    #  
plt.show()
    #  
```

```python
# x,y,w,h      
# x    x = 437
# y    y = 89
# w       = 187
# h       = 187
refined_x = x - w // 2 # 437 - (187//2) = 437-93 = 344
refined_y = y - h # 89-187 = -98
#       x, y  .   top-left  .
# , refined_x, refined_y   
print ('(x,y) : (%d,%d)'%(refined_x, refined_y)) #   :   
#       ,  -         
# opencv ndarray , ndarray        .
```

```python
# sticker_area     crop  .
#  (344,0)  (344+187, 0+89)   .
#   y,x . (y,x,rgb channel)
# img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
# img_show[0:0+89, 344:344+187]
# img_show[0:89, 344:531]
# , x 344~531 / y 0~89 
sticker_area = img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
#     0     .
#   0    ,    .
#       ------  
#   0 (  )  
# sticker_area        .
# ,      (   ------ )     sticker_area(       ) ,
#     
# np.where       .
#   img_sticker 0 (      ) sticker_area(       ) ,
#   img_sticker( ) .
img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==0,sticker_area,img_sticker).astype(np.uint8)
print("~")
```
